## 1. Importing Dependecies & Loading the Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, uniform, norm, expon
```

```python
customer_df = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
customer_df.head()
```

## 2. Selecting the Numerical Variables

```python
customer_num = customer_df.select_dtypes(include='number')
customer_num
```

## 3. Vizualizations of Distributions

```python
for variables in customer_num:
    plt.figure(figsize=(16,6))
    sns.displot(data=customer_df, x=customer_df[variables], kde=True)
    plt.show()
```

**Comment:** It seems that none of our numerical variables present a normal distribution.

## 4. Transformations of Data

### 4.1 Log Transformation

```python
customer_num_log = np.log(customer_num)
```

```python
for variables in customer_num_log:
    plt.figure(figsize=(16,6))
    sns.displot(data=customer_num_log, x=customer_num_log[variables], kde=True)
    plt.show()
```

**Comment:** I wouldn't necessarily call any of log-transformed distributions Gaussian per say but we do notice that it looks closer to a normal distribution.

### 4.2. Square Root Transformation

```python
customer_num_sqrt = np.sqrt(customer_num)
```

```python
for variables in customer_num_sqrt:
    plt.figure(figsize=(16,6))
    sns.displot(data=customer_num_sqrt, x=customer_num_sqrt[variables], kde=True)
    plt.show()
```

**Comment:** The square root transformed data is clearly more normally distributed than the original data. A great example of that is the Total Claim Amount which now almost looks like a Gaussian curve.

### 4.3 Cube Root Transformation

```python
customer_num_cbrt = np.cbrt(customer_num)
```

```python
for variables in customer_num_cbrt:
    plt.figure(figsize=(16,6))
    sns.displot(data=customer_num_cbrt, x=customer_num_cbrt[variables], kde=True)
    plt.show()
```

**Comment:** The cube root transformed data is much more normally distributed than the original data. A great example of that is,  again, the Total Claim Amount which now almost looks like a Gaussian curve.

### 4.4 Box Cox Transformation

For this transformation data needs to be 1 dimensional since the box-cox transformation produces an array. Let's take only the Total Claim Amount as an example and make two plots (one pre- and one post-transformation).

```python
from scipy import stats

customer_num_boxcox = stats.boxcox(customer_num['Total Claim Amount'])
customer_num_boxcox
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
x = customer_num['Total Claim Amount']
prob = stats.probplot(x, dist=stats.norm, plot=ax1)
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
plt.show()
```

```python
fig = plt.figure()
ax2 = fig.add_subplot(212)
xt, _ = stats.boxcox(x)
prob = stats.probplot(xt, dist=stats.norm, plot=ax2)
ax2.set_title('Probplot after Box-Cox transformation')
plt.show()
```

**Comment:** Using a probplot we see that the box-cox transformed data is much more normally distributed.
